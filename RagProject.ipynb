```python
from llama_index import VectorStoreIndex,SimpleDirectoryReader,ServiceContext
from llama_index.llms import HuggingFaceLLM
from llama_index.prompts.prompts import SimpleInputPrompt
```

```python
!mkdir Data
```

```python
document=SimpleDirectoryReader("/content/Data").load_data()
```

```python
document
```

```python
system_prompt="""
You are a Q&A assistant.Your goal is to answer questions as accurately as possible based on the instruction and the content provided
"""
query_wrapper_prompt=SimpleInputPrompt("<|USER|>(query_str)<|ASSISTANT|>")
```

```python
!huggingface-cli login
```

```python
from re import template
##Loading Mistral model
import torch
llm=HuggingFaceLLM(
    context_window=4096,
    max_new_tokens=256,
    generate_kwargs={"temparature":0.0,"db_sample":False},
    system_prompt=system_prompt,
    query_wrapper_prompt=query_wrapper_prompt,
    tokenizer_name="tiiuae/falcon-7b",
    model_name="tiiuae/falcon-7b",
    model_kwargs={"torch_dtype":torch.float16,"load_in_8bit":True}
)
```

```python
from langchain.embeddings.huggingface import HuggingFaceBgeEmbeddings
from llama_index import ServiceContext
from llama_index.embeddings import LangchainEmbedding
```

```python
embed_model=LangchainEmbedding(
    HuggingFaceBgeEmbeddings(model_name="sentence-transformers/all-mpnet-base-v2")
)
```

```python
service_context=ServiceContext.from_defaults(
    chunk_size=1024,
    llm=llm,
    embed_model=embed_model
)
```

```python
index=VectorStoreIndex.from_documents(document,service_context=service_context)
```

```python
query_engine=index.as_query_engine()
```

```python
query="What is dedicated team sequence?"
response=query_engine.query(query)
print(response)
```
